```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
credit_df = pd.read_csv('credit_dataset_final.csv')
credit_df.head()
```

```python
credit_df.info()
```

```python
credit_df.describe()
```

```python
credit_df.isnull().sum().sort_values(ascending = False)
```

```python
x_data = credit_df.drop(['credit.rating'], axis = 1)
y_data = pd.DataFrame(credit_df['credit.rating'])
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, random_state = 0)
train_df = pd.concat((x_train, y_train), axis = 1)
print('Tamanho do dataset de treino: {}, Tamanho do dataset de test: {}'.format(len(x_train), len(x_test)))
```

```python
train_df.head()
```

```python
plt.figure(figsize = (20, 20))
corimg = sns.heatmap(train_df.corr(), annot = True)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
