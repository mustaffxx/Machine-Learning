```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
train['train'] = 1
test['train'] = 0
all_data = pd.concat([train, test])
```

```python
all_data.info()
```

```python
all_data[all_data.columns[all_data.isnull().any() > 0]].isnull().sum().sort_values(ascending = False)
```

```python
all_data.head()
```

```python
all_data.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu'], axis = 1, inplace = True)
```

```python
all_data = pd.get_dummies(all_data, drop_first = True)
all_data.head()
```

```python
from sklearn.impute import SimpleImputer

original_data = all_data.copy()
my_imputer = SimpleImputer(strategy = 'most_frequent')
new_data = pd.DataFrame(my_imputer.fit_transform(all_data))
new_data.columns = original_data.columns
```

```python
new_data.head()
```

```python
new_train = new_data[new_data['train'] == 1]
new_test = new_data[new_data['train'] == 0]
new_train.drop(['train'], axis = 1, inplace = True)
new_test.drop(['train'], axis = 1, inplace = True)
new_test.drop(['SalePrice'], axis = 1, inplace = True)
```

```python
new_train_y = new_train['SalePrice']
new_train.drop(['SalePrice'], axis = 1, inplace = True)
x_train, x_val, y_train, y_val = train_test_split(new_train, new_train_y, random_state = 5)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

xgb_model = XGBRegressor()
xgb_model.fit(x_train, y_train)
xgb_model.score(x_val, y_val)
```

```python
best_xgb_model = XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=0.6, gamma=0,
             importance_type='gain', learning_rate=0.1, max_delta_step=0,
             max_depth=3, min_child_weight=1.5, missing=None, n_estimators=5000,
             n_jobs=6, nthread=-1, objective='reg:linear', random_state=5,
             reg_alpha=1e-05, reg_lambda=1, scale_pos_weight=1, seed=None,
             silent=None, subsample=0.95, verbosity=1)

best_xgb_model.fit(x_train, y_train)
best_xgb_model.score(x_val, y_val)
```

```python
submission = pd.DataFrame()
submission['Id'] = test['Id']
submission['SalePrice'] = best_xgb_model.predict(new_test)
submission.to_csv('submission.csv', index = False)
```
